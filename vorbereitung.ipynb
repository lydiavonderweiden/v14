```python
def cubes_to_vector(cube_numbers, diagonal=False):
    TOTAL_CUBES = 9
    vector = np.zeros(TOTAL_CUBES)
    for c in cube_numbers:
        vector[c-1] = 1
    if diagonal:
        vector *= np.sqrt(2)
    return vector
```

```python
rows_linear = [
    [1,2,3],
    [4,5,6],
    [7,8,9],
    #[1,4,7],
    [2,5,8],
    [3,6,9]
]
rows_diagonal = [
    [1,5,9],
    [3,5,7],
    #[4,8],
    [2,6],
    #[2,4],
    [6,8],
    [7,5,3]
]
a = (list(cubes_to_vector(r) for r in rows_linear) + 
list(cubes_to_vector(r,diagonal=True) for r in rows_diagonal))

a = np.array(a)
a
```
